```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException, TimeoutException, StaleElementReferenceException
import time
import pandas as pd
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url = 'https://www.olx.co.id/bandung-kota_g4000018/dijual-rumah-apartemen_c5158?filter=type_eq_rumah'
driver.get(url)
```

```python
while True:
    time.sleep(1.5)
    try:
        more_button = driver.find_element(By.XPATH, '//*[@data-aut-id="btnLoadMore"]')
        more_button.click()
    except NoSuchElementException:
        break

items = driver.find_elements(By.CLASS_NAME, '_1DNjI')

link_list = []
for item in items:
    # links = item.find_element(By.XPATH, '//*[@id="container"]/main/div/div/section/div/div/div[5]/div[2]/div/div[2]/ul/li[2]/a').text
    links = item.find_element(By.TAG_NAME, 'a')
    link = links.get_attribute('href')
    link_list.append(link)
    # print(link)

df = pd.DataFrame(link_list, columns= ['Links'])
print(len(df['Links']))
```

```python
while True:
    '''
    While loop to load all results for the search query by finding and clicking on the "Load More" button unless
    the button doesn't exist i.e., all results loaded, no more results to load. Without this, only the first 20 listings
    will be scraped
    '''
    try:
        WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, '//*[@data-aut-id="btnLoadMore"]')))
        driver.find_element(By.XPATH, '//*[@data-aut-id="btnLoadMore"]').click()
        print("Clicked 'Load More'")
    except ( NoSuchElementException, TimeoutException, StaleElementReferenceException ):
        break
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
complete = []
count = 0
for link in df["Links"]:
    try:
        #     time.sleep(1.5)
        driver.get(link)
        #     details = driver.find_element(By.CLASS_NAME, '_3nSm3').text
        price = driver.find_element(By.CLASS_NAME, "T8y-z").text
        loc = driver.find_element(By.CLASS_NAME, "_1RkZP").text
        try:
            # details = driver.find_element(By.XPATH, '//*[@data-aut-id="itemParams"]').text
            type = driver.find_element(By.XPATH, '//*[@data-aut-id="value_type"]').text
        except (
            NoSuchElementException,
            TimeoutException,
            StaleElementReferenceException,
        ):
            type = ""
            pass
        try:

            sqr_building = driver.find_element(
                By.XPATH, '//*[@data-aut-id="value_p_sqr_building"]'
            ).text
        except (
            NoSuchElementException,
            TimeoutException,
            StaleElementReferenceException,
        ):
            sqr_building = ""
            pass
        try:

            sqr_land = driver.find_element(
                By.XPATH, '//*[@data-aut-id="value_p_sqr_land"]'
            ).text
        except (
            NoSuchElementException,
            TimeoutException,
            StaleElementReferenceException,
        ):
            sqr_land = ""
            pass
        try:

            bedroom = driver.find_element(
                By.XPATH, '//*[@data-aut-id="value_p_bedroom"]'
            ).text
        except (
            NoSuchElementException,
            TimeoutException,
            StaleElementReferenceException,
        ):
            bedroom = ""
            pass
        try:

            bathroom = driver.find_element(
                By.XPATH, '//*[@data-aut-id="value_p_bathroom"]'
            ).text
        except (
            NoSuchElementException,
            TimeoutException,
            StaleElementReferenceException,
        ):
            bathroom = ""
            pass
        try:

            floor = driver.find_element(
                By.XPATH, '//*[@data-aut-id="value_p_floor"]'
            ).text
        except (
            NoSuchElementException,
            TimeoutException,
            StaleElementReferenceException,
        ):
            floor = ""
            pass
        try:

            facility = driver.find_element(By.CLASS_NAME, "_1qxGU").text
        except (
            NoSuchElementException,
            TimeoutException,
            StaleElementReferenceException,
        ):
            facility = ""
            pass
        try:

            certificate = driver.find_element(
                By.XPATH, '//*[@data-aut-id="value_p_certificate"]'
            ).text
        except (
            NoSuchElementException,
            TimeoutException,
            StaleElementReferenceException,
        ):
            certificate = ""
            pass
        # try :
        #     # details = driver.find_element(By.XPATH, '//*[@data-aut-id="itemParams"]').text
        #     desc = driver.find_element(By.XPATH, '//*[@data-aut-id="itemDescriptionContent"]').text
        # except ( NoSuchElementException, TimeoutException, StaleElementReferenceException ):
        #     dec = ''
        #     pass
        try:
            # details = driver.find_element(By.XPATH, '//*[@data-aut-id="itemParams"]').text
            date = driver.find_element(
                By.XPATH, '//*[@data-aut-id="itemCreationDate"]'
            ).text
        except (
            NoSuchElementException,
            TimeoutException,
            StaleElementReferenceException,
        ):
            date = ""
            pass

        final = {
            "link": link,
            "price": price,
            # 'Details' : details,
            "sqr_building": sqr_building,
            "sqr_land": sqr_land,
            "bedroom": bedroom,
            "bathroom": bathroom,
            "floor": floor,
            "facility": facility,
            "certificate": certificate,
            "date": date,
            "location": loc,
        }
        complete.append(final)
        count += 1
        print(f'Scrape : {count} out of {len(df["Links"])} {(count/len(df["Links"]))*100}', end='\r')
    except (NoSuchElementException, TimeoutException, StaleElementReferenceException):
        print(f"Skipped : {link}")
        pass


driver.close()

print(len(complete))
```
